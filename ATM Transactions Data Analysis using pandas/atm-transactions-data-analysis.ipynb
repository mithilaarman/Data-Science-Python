```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# dataset
file_path_new = '/kaggle/input/atm-transactions/atm_transactions.csv'
data_new = pd.read_csv(file_path_new)
data_new.info()
```

# ***Distribution Analysis***

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(data_new['totalBalance'], kde=True)
plt.show()
```

# ***Correlation Analysis***

```python
# only numeric columns for correlation analysis
numeric_data = data_new.select_dtypes(include=['float64', 'int64'])

# calculate the correlation matrix
correlation_matrix = numeric_data.corr()
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

# ***Transaction Volume Analysis Results***

```python
# convert transactionTime to datetime
data_new['transactionTime'] = pd.to_datetime(data_new['transactionTime'])

# extracting hour, date, day of the week, and week number from transactionTime
data_new['hour'] = data_new['transactionTime'].dt.hour
data_new['date'] = data_new['transactionTime'].dt.date
data_new['day_of_week'] = data_new['transactionTime'].dt.day_name()
data_new['week'] = data_new['transactionTime'].dt.isocalendar().week

# print to check data
print("Sample data:")
print(data_new.head())

# by hour to see hourly trends
hourly_transactions = data_new.groupby('hour').agg({
    'totalNumberTransaction': 'sum'
})
print("\nHourly Transactions:")
print(hourly_transactions.head())

# by day of the week to see daily trends
daily_transactions = data_new.groupby('day_of_week').agg({
    'totalNumberTransaction': 'sum'
}).reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
print("\nDaily Transactions:")
print(daily_transactions.head())

# by week to see weekly trends
weekly_transactions = data_new.groupby('week').agg({
    'totalNumberTransaction': 'sum'
})
print("\nWeekly Transactions:")
print(weekly_transactions.head())

# plotting the hourly, daily, and weekly
fig, ax = plt.subplots(3, 1, figsize=(14, 18))

ax[0].bar(hourly_transactions.index, hourly_transactions['totalNumberTransaction'])
ax[0].set_title('Hourly Transaction Volume')
ax[0].set_xlabel('Hour of the Day')
ax[0].set_ylabel('Total Number of Transactions')

ax[1].bar(daily_transactions.index, daily_transactions['totalNumberTransaction'], color='green')
ax[1].set_title('Daily Transaction Volume')
ax[1].set_xlabel('Day of the Week')
ax[1].set_ylabel('Total Number of Transactions')

ax[2].plot(weekly_transactions.index, weekly_transactions['totalNumberTransaction'], marker='o', color='red')
ax[2].set_title('Weekly Transaction Volume')
ax[2].set_xlabel('Week Number')
ax[2].set_ylabel('Total Number of Transactions')

plt.tight_layout()
plt.show()
```
